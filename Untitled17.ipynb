```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Generate date range
start_date = datetime.strptime("2022-01-01", "%Y-%m-%d")
dates = [start_date + timedelta(days=i) for i in range(2000)]

# Farmers (Wheat Surplus)
farmers_data = {
    "date": dates,
    "crop_type": ["Wheat"] * 2000,
    "quantity_kg": np.random.normal(loc=500, scale=100, size=2000).round(2)
}
farmers_df = pd.DataFrame(farmers_data)
farmers_df.to_csv("farmers.csv", index=False)

# Retailers (Dairy Surplus)
retailers_data = {
    "date": dates,
    "product_type": ["Dairy"] * 2000,
    "quantity_kg": np.random.normal(loc=300, scale=50, size=2000).round(2)
}
retailers_df = pd.DataFrame(retailers_data)
retailers_df.to_csv("retailers.csv", index=False)

# NGOs (Grains Demand)
ngos_data = {
    "date": dates,
    "food_type": ["Grains"] * 2000,
    "quantity_needed_kg": np.random.normal(loc=400, scale=70, size=2000).round(2)
}
ngos_df = pd.DataFrame(ngos_data)
ngos_df.to_csv("ngos.csv", index=False)

# Covariates (Temperature, Market Price)
covariates_data = {
    "date": dates,
    "temperature": np.random.normal(loc=30, scale=5, size=2000).round(2),
    "market_price": np.random.normal(loc=50, scale=10, size=2000).round(2)
}
covariates_df = pd.DataFrame(covariates_data)
covariates_df.to_csv("covariates.csv", index=False)

print("✅ CSVs generated: farmers.csv, retailers.csv, ngos.csv, covariates.csv")
```

```python
import pandas as pd
from darts import TimeSeries
from darts.models import TiDEModel
from darts.dataprocessing.transformers import Scaler
from darts.metrics import mape

# CSV/Excel source (assuming you've loaded them already)
def load_data():
    farmers_df = pd.read_csv("farmers.csv")  # Replace with actual data source
    retailers_df = pd.read_csv("retailers.csv")
    ngos_df = pd.read_csv("ngos.csv")
    covariates_df = pd.read_csv("covariates.csv")
    return farmers_df, retailers_df, ngos_df, covariates_df

# Train DARTS models and generate forecasts
def train_and_forecast():
    # Load data
    farmers_df, retailers_df, ngos_df, covariates_df = load_data()

    # Convert to DARTS TimeSeries
    farmers_ts = TimeSeries.from_dataframe(farmers_df, time_col='date', value_cols=['quantity_kg'])
    retailers_ts = TimeSeries.from_dataframe(retailers_df, time_col='date', value_cols=['quantity_kg'])
    ngos_ts = TimeSeries.from_dataframe(ngos_df, time_col='date', value_cols=['quantity_needed_kg'])
    covariates_ts = TimeSeries.from_dataframe(covariates_df, time_col='date', value_cols=['temperature', 'market_price'])

    # Split data
    split_idx = int(0.8 * len(farmers_ts))
    train_farmers, test_farmers = farmers_ts[:split_idx], farmers_ts[split_idx:]
    train_retailers, test_retailers = retailers_ts[:split_idx], retailers_ts[split_idx:]
    train_ngos, test_ngos = ngos_ts[:split_idx], ngos_ts[split_idx:]
    train_covariates = covariates_ts[:split_idx]

    # Scale data
    scaler = Scaler()
    train_farmers_scaled = scaler.fit_transform(train_farmers)
    train_retailers_scaled = scaler.fit_transform(train_retailers)
    train_ngos_scaled = scaler.fit_transform(train_ngos)

    # Initialize and train models
    model_farmers = TiDEModel(input_chunk_length=30, output_chunk_length=7, n_epochs=10)
    model_retailers = TiDEModel(input_chunk_length=30, output_chunk_length=7, n_epochs=10)
    model_ngos = TiDEModel(input_chunk_length=30, output_chunk_length=7, n_epochs=10)

    model_farmers.fit(train_farmers_scaled, past_covariates=train_covariates)
    model_retailers.fit(train_retailers_scaled, past_covariates=train_covariates)
    model_ngos.fit(train_ngos_scaled, past_covariates=train_covariates)

    # Generate forecasts
    forecast_farmers = scaler.inverse_transform(model_farmers.predict(n=7, past_covariates=covariates_ts))
    forecast_retailers = scaler.inverse_transform(model_retailers.predict(n=7, past_covariates=covariates_ts))
    forecast_ngos = scaler.inverse_transform(model_ngos.predict(n=7, past_covariates=covariates_ts))

    # Evaluate
    print(f"MAPE Farmers: {mape(test_farmers[:7], forecast_farmers):.2f}%")
    print(f"MAPE Retailers: {mape(test_retailers[:7], forecast_retailers):.2f}%")
    print(f"MAPE NGOs: {mape(test_ngos[:7], forecast_ngos):.2f}%")

    # Prepare forecast DataFrames
    df_farmers = pd.DataFrame({
        'entity': 'farmers',
        'date': forecast_farmers.time_index.format(),
        'quantity': forecast_farmers.values().flatten(),
        'forecast_type': 'surplus'
    })

    df_retailers = pd.DataFrame({
        'entity': 'retailers',
        'date': forecast_retailers.time_index.format(),
        'quantity': forecast_retailers.values().flatten(),
        'forecast_type': 'surplus'
    })

    df_ngos = pd.DataFrame({
        'entity': 'ngos',
        'date': forecast_ngos.time_index.format(),
        'quantity': forecast_ngos.values().flatten(),
        'forecast_type': 'demand'
    })

    # Save to Excel
    final_df = pd.concat([df_farmers, df_retailers, df_ngos])
    final_df.to_excel("forecasts.xlsx", index=False)

    # Print sample forecasts
    print("\nSample Forecasts:")
    print("Farmers Surplus (Wheat):")
    print(df_farmers.head(3))
    print("Retailers Surplus (Dairy):")
    print(df_retailers.head(3))
    print("NGOs Demand (Grains):")
    print(df_ngos.head(3))

# Main execution
if __name__ == '__main__':
    train_and_forecast()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Dummy data for demonstration (remove this if you already have actual data)
# Replace this with your actual test and forecast data
np.random.seed(0)
test_supply = pd.Series(np.random.randint(50, 150, size=10))
forecast_supply = pd.Series(np.random.randint(50, 150, size=10))
test_demand = pd.Series(np.random.randint(50, 150, size=10))
forecast_demand = pd.Series(np.random.randint(50, 150, size=10))

# Convert to binary classes using threshold
threshold = 100

# Extract first 7 days and convert to binary (1 if > threshold, else 0)
y_true_supply = (test_supply[:7].values > threshold).astype(int)
y_pred_supply = (forecast_supply[:7].values > threshold).astype(int)

y_true_demand = (test_demand[:7].values > threshold).astype(int)
y_pred_demand = (forecast_demand[:7].values > threshold).astype(int)

# Metrics for supply
print("\n📦 Supply Forecast Classification Metrics:")
print(f"Accuracy:  {accuracy_score(y_true_supply, y_pred_supply):.2f}")
print(f"Precision: {precision_score(y_true_supply, y_pred_supply, zero_division=0):.2f}")
print(f"Recall:    {recall_score(y_true_supply, y_pred_supply, zero_division=0):.2f}")
print(f"F1 Score:  {f1_score(y_true_supply, y_pred_supply, zero_division=0):.2f}")

# Metrics for demand
print("\n🍽 Demand Forecast Classification Metrics:")
print(f"Accuracy:  {accuracy_score(y_true_demand, y_pred_demand):.2f}")
print(f"Precision: {precision_score(y_true_demand, y_pred_demand, zero_division=0):.2f}")
print(f"Recall:    {recall_score(y_true_demand, y_pred_demand, zero_division=0):.2f}")
print(f"F1 Score:  {f1_score(y_true_demand, y_pred_demand, zero_division=0):.2f}")
```
